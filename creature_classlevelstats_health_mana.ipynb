# *Estimate Health/Mana for 63+ creatures*

We want to calculate `health` and `mana` values from table `creature_classlevelstats` for creatures above `level 63` and for each `unit_class`.

```python
import matplotlib.pyplot as plt
```

## Health Func

- Formula : `new_health = health + (10 + new_level) * class_multiplier`
- Mostly off under level 20, since we want to calculate high level health, it doesn't matter

```python
CLASS_HEALTH_MUTIPLIERS = {1: 1.275, 2: 1.0155, 8: 0.888 }


# Get health by level
def get_health(lvl, unit_class=1):
  class_health_lvl_1 = {1: 42, 2: 41, 8: 40}
  multiplier = CLASS_HEALTH_MUTIPLIERS.get(unit_class, 1)
  health = class_health_lvl_1.get(unit_class, 1)

  for lvl in range(1, lvl):
    health += ( 10 + lvl ) * multiplier

  return int(health)

# Get next level health - more accurate than get_health if current_health is blizzlike
def get_next_lvl_health(current_health, current_lvl, unit_class=1):
  multiplier = CLASS_HEALTH_MUTIPLIERS.get(unit_class, 1)

  return int(current_health + (10 + current_lvl + 1) * multiplier)
```

## Health Estimation

```python
print("#### LEVEL 30 #####")

lvl_30_health = get_health(30)
print("get_health estimation          : ", lvl_30_health)
lvl_30_health2 = get_next_lvl_health(905, 29)
print("get_next_lvl_health estimation : ", lvl_30_health2)
print("blizzlike data                 : ", 955)

print("\n\n")

print("#### LEVEL 40 #####")

lvl_40_health = get_health(40)
print("get_health estimation          : ", lvl_40_health)
lvl_40_health2 = get_next_lvl_health(1459, 39)
print("get_next_lvl_health estimation : ", lvl_40_health2)
print("blizzlike data                 : ", 1524)

print("\n\n")

print("#### LEVEL 50 #####")

lvl_50_health = get_health(50)
print("get_health estimation          : ", lvl_50_health)
lvl_50_health2 = get_next_lvl_health(2138, 49)
print("get_next_lvl_health estimation : ", lvl_50_health2)
print("blizzlike data                 : ", 2215)

print("\n\n")

print("#### LEVEL 60 #####")

lvl_60_health = get_health(60)
print("get_health estimation          : ", lvl_60_health)
lvl_60_health2 = get_next_lvl_health(2961, 59)
print("get_next_lvl_health estimation : ", lvl_60_health2)
print("blizzlike data                 : ", 3052)
```

## Mana Func

- Formula : `new_mana = mana + (10 + new_level) * class_multiplier`
- Mostly off under level 20, since we want to calculate high level mana, it doesn't matter

```python
CLASS_MANA_MULTIPLIERS = {2: 1.03, 8: 2.395 }

# Get mana by level
def get_mana(lvl, unit_class=2):
  class_mana_level_1 = {2: 60, 8: 120}
  mana = class_mana_level_1.get(unit_class, 2)
  multiplier = CLASS_MANA_MULTIPLIERS.get(unit_class, 2)

  for lvl in range(1, lvl):
    mana += ( 10 + lvl ) * multiplier

  return int(mana)

# Get next level mana, more accurate than get_mana if current_mana is blizzlike
def get_next_lvl_mana(current_mana, current_lvl, unit_class=2):
  multiplier = CLASS_MANA_MULTIPLIERS.get(unit_class, 2)

  return int(current_mana + (10 + current_lvl+1) * multiplier)
```

## Mana Estimation

```python
print("#### LEVEL 30 #####")

lvl_30_mana = get_mana(30)
print("get_mana estimation          : ", lvl_30_mana)
lvl_30_mana2 = get_next_lvl_health(811, 29)
print("get_next_lvl_mana estimation : ", lvl_30_mana2)
print("blizzlike data               : ", 852)

print("\n\n")

print("#### LEVEL 40 #####")

lvl_40_mana = get_mana(40)
print("get_mana estimation          : ", lvl_40_mana)
lvl_40_mana2 = get_next_lvl_mana(1236, 39)
print("get_next_lvl_mana estimation : ", lvl_40_mana2)
print("blizzlike data               : ", 1283)

print("\n\n")

print("#### LEVEL 50 #####")

lvl_50_mana = get_mana(50)
print("get_mana estimation          : ", lvl_50_mana)
lvl_50_mana2 = get_next_lvl_mana(1750, 49)
print("get_next_lvl_mana estimation : ", lvl_50_mana2)
print("blizzlike data               : ", 1807)

print("\n\n")

print("#### LEVEL 60 #####")

lvl_60_mana = get_mana(60)
print("get_mana estimation          : ", lvl_60_mana)
lvl_60_mana2 = get_next_lvl_mana(2369, 59)
print("get_next_lvl_mana estimation : ", lvl_60_mana2)
print("blizzlike data               : ", 2434)
```
